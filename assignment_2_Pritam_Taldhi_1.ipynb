# CALCULATING MEAN

```python
def mean_type(x, y, size):
    b=np.zeros(4)
    l=np.ndarray(shape=(size,4))
    for i in range(size):
        b += img_arr[x[i],y[i],:]

    b /= size

    for i in range(size):
        l[i] = np.subtract(img_arr[x[i],y[i],:],b)

    return (b, l)
```

# CALCULATING THE  COVARIANCE MATRIX

```python
def cov_type(m, size):
    cov = np.ndarray(shape=(4, 4), dtype=np.float64)
    for i in range(4):
        for j in range(4):
            cov[i][j]=np.dot(m[:,i],m[:,j])/size
    return cov
```

```python
riv_cov = cov_type(riv_vec, 50); riv_cov
```

```python
non_cov = cov_type(non_vec, 150); non_cov
```

# CALCULATING THE INVERSE OF THE COVARIANCE MATRIX

```python
riv_inv = np.linalg.inv(riv_cov); riv_inv
```

```python
non_inv = np.linalg.inv(non_cov); non_inv
```

# DEFINING THE NAIVE BAYES CLASSIFIER

```python
def cal(i,j):
    r = np.subtract(img_arr[i,j,:],riv_me)
    nr = np.subtract(img_arr[i,j,:],non_me)

    river_class = np.dot(np.dot(r.T,riv_inv), r)
    non_river_class = np.dot(np.dot(nr.T,non_inv), nr)
    
    det_cov_r = np.linalg.det(riv_cov)
    p1 = (-0.5) * 1/np.sqrt(det_cov_r) * np.exp(river_class);

    det_cov_nr = np.linalg.det(non_cov)
    p2 = (-0.5) * 1/np.sqrt(det_cov_nr) * np.exp(non_river_class);
    return p1,p2
```

```python
def bayes(b1, b2):
    output = np.ndarray(shape=(512, 512), dtype = np.int64)
    for i in range(512):
        for j in range(512):
            p1,p2 = cal(i,j)
            if((b1 * p1) <= (b2 * p2)):
                output[i,j]=255
            else:
                output[i,j]=0
                
    return output
```

# CALCULATING THE ACCURACY

```python
def accuracy(sample, b1, b2):   
    im = bayes(b1,b2)

    count=0
    for i in range(512):
        for j in range(512):
            if sample[i,j, 3]==im[i,j]:
                count = count+1
    print(f'Accuracy is {count*100/(512*512)}%')
    # the plot
    plt.subplot(1,1,1)
    plt.imshow(im, cmap='gray')
    plt.tick_params(labelleft=False, labelbottom=False, bottom=False,top=False,right=False,left=False, which='both')
    plt.subplots_adjust(right=1.2, top=2.5)
    plt.show()
```
